## Relation between functions, classes and objects

A function is basicly a callable object of a class. It is an **abstraction**, and basicly an easier way to write and use code. 

```python
def my_func():
    return 'From my_func'
```

```python
print(type(my_func))
```

```python
class MyClass:
    pass
```

```python
print(type(MyClass))
```

### How to write a function through a class? 

You make the object of that class callable

```python
my_func()
```

```python
class MyClass:
    def __call__(self):
        return 'From call function'
```

```python
my_func = MyClass()
```

```python
my_func()
```

This means that anything you would define in a function could be definned in a classÂ´s \_\_call\_\_ method.

### Compute

```python
def compute():
    l = []
    for i in range(10):
        l.append(i)
    return l
```

```python
compute()
```

```python
class Compute:
    def __call__(self):
        l = []
        for i in range(10):
            l.append(i)
        return l
```

```python
compute = Compute()
```

```python
compute()
```

## Generators

```python
from time import sleep
```

```python
def compute():
    l = []
    for i in range(10):
        sleep(.5)
        l.append(i)
    return l
```

```python
compute()
```

```python
class Compute:
    def __call__(self):
        l = []
        for i in range(10):
            sleep(.5)
            l.append(i)
        return l
```

```python
compute = Compute()
```

```python
compute()
```

### Iterator class

```python
class Compute:
    def __call__(self):
        return iter(self)
    def __iter__(self):
        self.last = 0
        return self
    def __next__(self):
        if self.last > 10:
            raise StopIteration
        self.last += 1
        return self.last
```

```python
compute = Compute()
```

```python
compute()
```

```python
#it = iter(compute)
it = compute()
```

```python
next(it)
```

### generator function

```python
def compute():
    for i in range(10):
        yield i
```

```python
compute()
```

```python
it = compute()
```

```python
next(it)
```

### Generator expression

```python
it = (i for i in range(10))
```

```python
next(it)
```

### for loop uses the iterator protocol

```python
for i in it:
    print(i)
```
